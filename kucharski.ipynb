```python
Z_values = []
k_values = []
for z in range(1, 20 + 1):
    Z_values.extend([z] * z)
    k_values.extend(range(1, z + 1))
df = pd.DataFrame({'Z': Z_values, 'k': k_values})
```

```python
# parameters
mu = 3.02
sigma = 0.29
tau1 = 0
tau2 = 0.06
```

```python
#cross reactivity
df['d'] =  exp(-sigma*abs((df['k']-df['Z'])) )
#S1 (boosting)
df['S1'] = (1+tau1)**(df['Z']-df['k'])
#S2 (supperesion)
df['S2'] = exp(-tau2*(df['k']-1))
# antibody titer
df['y'] = mu*df['S1']*df['S2']#*df['d']
```

```python
df_w = df.copy()
# parameters
mu_w = 5.01
sigma_w = 0.29
tau1_w = 0.05
tau2_w = 0.13
```

```python
#cross reactivity
df_w['d'] =  exp(-sigma_w*abs((df_w['k']-df_w['Z'])) )
#S1 (boosting)
df_w['S1'] = (1+tau1_w)**(df_w['Z']-df_w['k'])
#S2 (supperesion)
df_w['S2'] = exp(-tau2_w*(df_w['k']-1))
# antibody titer
df_w['y'] = mu_w*df_w['S1']*df_w['S2']#*df_w['d']
```
